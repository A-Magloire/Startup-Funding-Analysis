### Top 10 Private Equity Deals

```python
df_PI_Location = df_PI[["CityLocation", "AmountInUSD"]]
df_PI_Location.set_index("CityLocation", inplace=True)
```

```python
df_PI_Location_sum = df_PI_Location.groupby("CityLocation").sum()
df_PI_Location_sum.sort_values("AmountInUSD", ascending=False, inplace=True)
df_PI_Location_sum.reset_index(inplace=True)
df_PI_Location_sum
```

### Plot Indian Private Equity Investment Deals by City

```python
plt.figure(figsize=(20,7))
sns.barplot(data=df_PI_Location_sum,
            x=df_PI_Location_sum.CityLocation,
            y=df_PI_Location_sum["AmountInUSD"])
```

### Plot Indian Private Equity Investment Deals by Investment Vertical

```python
df_PI_Vertical = df_PI[["IndustryVertical", "AmountInUSD"]]
df_PI_Vertical_sum = df_PI_Vertical.groupby("IndustryVertical").sum()
```

```python
df_PI_Vertical_sum_top10 = df_PI_Vertical_sum.sort_values("AmountInUSD", ascending=False).head(10)
df_PI_Vertical_sum_top10.reset_index(inplace=True)
```

```python
plt.figure(figsize=(20,7))
sns.barplot(data=df_PI_Vertical_sum_top10,
            x=df_PI_Vertical_sum_top10["IndustryVertical"],
            y=df_PI_Vertical_sum_top10["AmountInUSD"])
```

```python
### Convert str to int
df1["AmountInUSD"].replace(",", "", regex=True, inplace=True)
df1["AmountInUSD"] = df1["AmountInUSD"].astype("float")
df1.head()
```

```python
df1_startups = df1[["StartupName", "AmountInUSD"]]
df1_top10_startups_sum = df1_startups.groupby("StartupName").sum()
df1_top10_startups_sum = df1_top10_startups_sum.sort_values("AmountInUSD", ascending=False).head(10)
df1_top10_startups_sum.reset_index(inplace=True)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(data=df1_top10_startups_sum, x=df1_top10_startups_sum["StartupName"], y=df1_top10_startups_sum["AmountInUSD"])
```
